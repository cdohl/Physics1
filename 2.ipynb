### 2.2 Order of magnitude

In physics to get a picture of the which quantity is large compared to others, order of magnitude estimages are conducted. Therefore each quantity is written in power of tens and then the calculation is done with these powers. Suitable rounding needs to be applied. Here are some examples:

$1.5\,$km $\rightarrow$ $10^3\,$m

$299792458\,$m/s $\rightarrow$ $10^8\,$m/s

*Example:*

Order of magnitude estimate the number of steps you walk in your lifetime. 
Say 1000 steps per day, 365 steps per year, 80 years

$10^3\,\text{steps} \cdot 10^2 \cdot 10^2=10^7\,$steps

Which distance do you walk in a lifetime?

$10^7\,\text{steps} \cdot 1\,$m/step$=10^7\,$m$=10^4\,$km.

### 2.3 Dimensional Analysis 

The following example is not an introduction into dimensional analysis, yet it should convince you with a simple example that physical laws can be deduced from the analysis of dimensions of the contributing quantities.

This example makes used of the fall time $t$ of an apple from a height $h$.
This fall time, $t$ may depend on 

* height $h$
* mass of the apple $m$
* gravitational acceleration $g$

Now we look at and compare the units of the quantities $h,m,\mathrm{und}\,g$ with the fall time $t$ using the knowledge that for a physical law the dimensions on the left and right-hand
 side of the equation must be the same, i.e. 

$$t\propto h^\alpha \, m^\beta \, g^\gamma \tag{2.1}$$

We have not to determine the unknown constants $\alpha, \beta, \gamma$ to meet this requirement through Eq. (2.1).

$$[\mathrm{T}^1]\propto [\mathrm{L}]^\alpha\,[\mathrm{M}]^\beta\,\frac{[\mathrm{L}]^\gamma}{[\mathrm{T}]^{2\gamma}}
$$


We reveal immediately that $\beta=0$ and $\alpha+\gamma=0$. For the unit of time $\alpha-2\gamma$
 must be one, i.e. $1=\alpha-2\gamma$. As a result, we obtain: 

$$\gamma=-1/2\quad  \mbox{and} \quad \alpha=1/2$$.

Now let us insert these constants into Eq. (2.1) we obtain: 

$$t=C \sqrt{\frac{h}{g}}\quad.\tag{2.2}$$

We do not know the constant $C$, yet we can already make the important statement that the fall time does not depend on the mass $m$ of the apple. 

### 2.4 Significant Figures


The significant figures (s.f.) of a measurements is the number of digits of the measurement. The measurement must start with a non-zero digit.

Examples: 

$1.256$ $\rightarrow$ $4$ s.f. 

$0.125$ $\rightarrow$ $3$ s.f. (because $1.25 \cdot 10^{-1}$)

**Rule 1:**

When multiplying or dividing multiple measurements, the number of significant figures in the final result is determined by the quantity with the fewest significant figures.

*Example:*

You measure a radius of $0.2\,$m of a circle. What is the area?

$A= \pi\, (0.2\,\text{m})^2=0.1\,$m$^2$

 
**Rule 2:**

When adding or subtracting measurements, the result should have the same number of decimal places as the term with the fewest decimal places.

*Example:*

$42.42 + 4.2 = 46.6$
