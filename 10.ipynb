# 10. Flowing liquids and gases (hydro- and aerodynamics)

## 10.1 Basic concepts

* Dynamics: 

  Sum of all momentum of the molecules 
  
  $$\sum \vec{p}_i\ne 0$$
  
  $\rightarrow$ flowing liquids/gases
  
* relative motion in relation to a fixed laboratory coordinate system

* motion is slow to the speed of sound (propagation speed of pressure changes)

* thermal motion is not taken into account
* Consideration of volume elements

  $$\Delta V = \frac{\Delta m}{\rho}$$

* Movement due to 
  * Pressure forces
  * gravitational force
  * frictional forces

## 10.2 Continuity equation

* Flow through a surface $\vec{A}$: the quantity $\vec{v}\cdot\vec{A}$ indicates the volume flow through the surface $\vec{A}$. The area is a vector, the magnitude indicates the size of the area and the vector the direction. 

<img src="figures/10.1_2.png" width=300px>

The volume flow $Q$ is 

$$Q=\vec{v}\cdot\vec{A}=\underbrace{|\vec{v}|\,|\vec{A}|\,\cos(\vec{v},\vec{A})}_{A_\perp}=|\vec{v}|\,A_\perp= v\,A$$


* Continuity equation

<img src="figures/10.1_3.png" width=420px>

$$\Delta V_1=A_1\Delta x_1\quad\quad\quad\Delta V_2=A_2\Delta x_2$$

* Conservation of the mass flow through the area $A_1$ or $A_2$

* The mass flow is

$$\frac{\Delta m}{\Delta t}$$

This means that $$\Delta m_1=\rho_1\,\Delta V_1$$
and $$\Delta m_2=\rho_2\,\Delta V_2$$


$$\frac{\Delta m_1}{\Delta t}=\frac{\Delta m_2}{\Delta t}$$

$$\rho_1\,A_1\,\frac{\Delta x_1}{\Delta t}=\rho_2\,A_2\,\frac{\Delta x_2}{\Delta t}$$

Border crossing $\Delta t \rightarrow 0$:

$$\rho_1\,A_1\,\frac{\mathrm{d} x_1}{\mathrm{d} t}=\rho_2\,A_2\,\frac{\mathrm{d} x_2}{\mathrm{d}t}$$

and with the definition for the oscillation we obtain the continuity equation for the mass flow:

$$\rho_1\,A_1\,v_1=\rho_2\,A_2\,v_2$$

If the density does not change, e.g. in incompressible liquids/gases, $\rho_1=\rho_2$ and the following applies

$$A_1\,v_1=A_2\,v_2=\mathrm{konst.}$$

## 10.3 Bernoulli equation


The force exerted on the segment by the fluid to the left of the blue portion in below Figure has a magnitude $F_1 = p_1 A_1$. During a time interval $\Delta t%$, the point of application  of  this  force  travels a distance of $\Delta x_1$,  as  the  blue  portion of fluid enters the section of pipe past point 1. The work done by this force is $W_1= F_1 \Delta x_1 =p_1 A_1 \Delta x_1 = p_1 V$, where $V$ is the volume  of the  blue  portion  of  fluid  passing  point  1.  

In  a  similar  manner, the work done by the segment by the fluid to the right of the segment in the  same  time  interval  $\Delta t$ is  $W_2 = - p_2 A_2 \Delta x_2 = p_2 V$,  where  $V$  is  the same volume as on the left. The negative sign comes from the fact that the work is done by the volume segment. 

<img src="figures/10.1_4.png" width=600px>

The  net  work  done  on  the  segment  by  these  forces  in  the  time  interval $\Delta t$ is then 

$$
W=(p_1-p_2) V
$$

This work changes the kinetic and potential energy of the fluid particles, 
thus 

$$
\Delta E=E_2-E_1=\frac{1}{2}\rho V v_2^2 + \rho V g y_2 -  \frac{1}{2}\rho V v_1^2 - \rho V g y_1=(p_1-p_2) V
$$

Now moving the energies and work at point 1 and point 2 on each side of the equation and dividing by the volume $V$ gives us:

$$
p_1+\frac{1}{2}\rho v_1^2 + \rho g y_1 = p_2+\frac{1}{2}\rho v_2^2 + \rho g y_2
$$

This equation is termed the Bernoulli equation, named after the Swiss physicist Daniel Bernoulli, who published the important book *Hydrodynamica* in 1738. If your are interested and can read Latin, here is the [link](https://www.e-rara.ch/zut/doi/10.3931/e-rara-3911) (chapter 12, or see [here](https://www.17centurymaths.com/contents/dbernoullicontents.htm) for the translation into English).

Bernoulli equation can be applied along a streamline in a steady (time independent) flow at two or more points. 


### Applications of Bernoulli Equation


#### Venturi Tube

<img src="figures/10.1_5.png" width=400px>


The  horizontal  constricted  pipe  as illustrated  in the  figure is named Venturi tube. By measuring the pressure of the flow at to points the liquid velocity can be be determined.

We apply Bernoulli equation for $y_1=y_2=0$, as the center streamline is on the same height.

$$
p_1 + \frac{1}{2} \rho v_1^2 = p_1 + \frac{1}{2} \rho v_1^2
$$

The ratio of the velocities because of conservation of mass is

$$v_1=\frac{A_2}{A_1}v_2\quad .$$

Insert this expression into Bernoulli equation and solve for $v_2$ gives us

$$
v_2=\sqrt{\frac{2(p_1-p_2)}{\rho(A_1^2-A_2^2)}}\quad .
$$


The pressure difference $\Delta p= p_1-p_2$ can be conveniently measured with a U-tube manomoeter.

#### Toricelli's law

An enclosed tank with a constant presure $p$ is emptied into the surrounding with constant pressure $p_0$ through a nozzle with an opening area of $A_1$. $A_1$ is much smaller than $A_2$, thus we can neglect the velocity at the free surface, i.e. $v_2\approx 0$.

<img src="figures/10.1_6.png" width=300px>

Applying Bernoulli between points 1 and 2 we obtain

$$
p_0+\frac{1}{2}\rho v_1^2 +\rho g y_1 =p+\rho g y_2 \quad .
$$

This can be solved for $v_1$ and we obtain

$$v_1=\sqrt{\frac{2(p-p_0)}{\rho}+2 g h}\quad .
$$
