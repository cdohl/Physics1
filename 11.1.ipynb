### 11.1 Free, undamped, harmonic oscillations
#### 11.1.1 Physical or compound pendulum

<img src="figures/24.1_2.png" alt="24.1_2" style="width: 200px;"/>

$$\vec{M}=J_A\,\vec{\alpha}$$

Here $M$ is the torque, and $J_A$ is the mass moment of inertia with respect to the suspension point $A$ and $\left|\vec{\alpha}\right|$ is the angular acceleration $\mathrm{d}^2\varphi/\mathrm{d}t^2$. The direction of $\vec{\alpha}$ indicates the axis of rotation. 

$$\vec{M}=\vec{r}\times\vec{F}=\vec{s}\times\vec{F}_g$$

$$\left| \vec{M}\right|=-\left|\vec{s}\right|\,\left|\vec{F}_g\right|\,\sin\varphi$$

The minus sign describes the torque against the angle $\varphi$, i.e. for positive angles $\varphi$ the amount of torque is negative and vice versa. If you solve the equation vectorially, you do not need this consideration, see for example [here](https://en.wikipedia.org/wiki/Pendulum#Compound_pendulum). 

$$-s \cdot m \cdot g \cdot \sin \varphi = J_A \, \frac{\mathrm{d}^2\varphi}{\mathrm{d}\,t^2}$$

$$\ddot{\varphi} + \frac{s\cdot m\cdot g}{J_A}\,\sin\varphi= 0\tag{11.1.1}$$

* for small deflection angles 
$\varphi\le 5^\circ$ ⇨ $\sin \varphi \approx \varphi$

and equation (11.1.1) simplifies to

$$\ddot{\varphi} + \frac{s\cdot m\cdot g}{J_A}\,\varphi= 0$$

#### 11.1.2 Spring oscillator

<img src="figures/24.1_3.png" alt="24.1_3" style="width: 400px;"/>

Spring force law:

$$\vec{F}_S=-k \cdot x\cdot \vec{e}_x$$

$$ \vec{F} = m\,\frac{\mathrm{d}^2 x}{\mathrm{d}\,t^2}\cdot \vec{e}_x=m\, \vec{a}$$

$$m\, \ddot{x} = - k \, x $$

or

$$\ddot{x}+\frac{k}{m}\, x = 0\tag{11.1.2}$$

This is a homogeneous 2nd order differential equation with constant coefficients.

#### 11.1.3 Solution of the differential equation: Spring oscillator
  
Which function $x=x(t)$ fulfills the differential equation (11.1.2)?
  
Approach: $x(t)=A\, e^{\,\lambda\,t}$
  
Now we have to determine the parameters $A$ and $\lambda$:
  
\begin{eqnarray}
x(t)&=&A\cdot e ^{\lambda t}\\
\dot{x}(t)&=&A\cdot \lambda\cdot e ^{\lambda t}\\\
\ddot{x}(t)&=&A\cdot \lambda^2\cdot e ^{\lambda t}
\end{eqnarray}

  ⇨ Insert into the ODE (11.1.2)
  

\begin{eqnarray}
A\cdot \lambda^2\cdot e ^{\lambda t} + \frac{k}{m} A\cdot e ^{\lambda t} &=&0\\\
A\cdot e ^{\lambda t}\cdot\left[\lambda^2 +\frac{k}{m}\right] &=&0
\end{eqnarray}
can only be fulfilled if $\left[\lambda^2 +\frac{k}{m}\right]=0$. This is called the characteristic equation and we determine its solutions.

$$\lambda^2=-\frac{k}{m}$$

We now denote $\frac{k}{m}=\omega_0^2$, the reason will soon become clear.

$$\lambda_{1,2}=\pm \sqrt{-\omega_0^2}=\pm\sqrt{-1}\,\omega_0=\pm i \omega_0$$

$$
\left.
\begin{array}{11}
\lambda_1=+ i \omega_0\\\
\lambda_2=- i \omega_0
\end{array}
\right\} \mathrm{two\,solutions}
$$


And insert into the ansatz

$$x(t)=A_1 e^{i\omega_0 t} +A_2 e^{-i\omega_0 t}\tag{11.1.3}$$

Determination of $A_1$ and $A_2$ from the initial conditions (IC) 

For the spring oscillation, $x(t=0)=x_0$ and $\dot{x}(t=0)=0$. The last IC states that the initial velocity is zero.


IC 1: $x(t=0)=x_0=A_1 + A_2$

IC 2: $\dot{x}=A_1\cdot i \omega_0\, e^{i\omega_0 t} - A_2\cdot i \omega_0\, e^{-i\omega_0 t}$
  
  for $t=0$: $\dot{x}=0$ ⇨ $A_1 i \omega_0 - A_2 i \omega_0=0$
   
   ⇨ $A_1=A_2$
   
   ⇨ With IC 1: $A_1=\frac{x_0}{2}$ und $A_2=\frac{x_0}{2}$
   
   
Inserting into the general solution Eq. (11.1.3):
   $$x(t)=\frac{x_0}{2}\underbrace{\left( e^{i\omega_0 t}+e^{-i\omega_0 t}\right)}_{2\cos(\omega_0 t)}$$

$$x(t)=\frac{x_0}{2}\cdot 2\cdot \cos( \omega_0 t)$$

<div class="alert alert-success">
$$x(t)=x_0\cdot \cos( \omega_0 t)$$
</div>

for $x(t=0)=x_0$ und $\dot{x}(t=0)=0$.


* Every other IC leads to different $A_1$ and $A_2$ ⇨ phase shifts.

In general:

<div class="alert alert-success">
$$x(t)=x_0 \, \cos\left(\omega_0 \cdot t - \varphi_0\right)$$
</div>

  $\varphi_0$ is the phase shift.

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi,500)

def update(x0=1., phi0 = 0.,omega0 = 1.):
    T0=2.*np.pi/omega0
    fig = plt.figure(figsize=(7,4))
    plt.plot(t/T0,x0* np.cos( omega0*t - phi0))
    plt.text(0,.75,r'$x(t)=x_0 \cos(\omega_0 \cdot t - \varphi_0)$',fontsize=15)
    plt.ylim(-1.,1.)
    plt.xlabel(r'$t/T_0$',fontsize=15)
    plt.ylabel('amplitude',fontsize=15)
    plt.show()

interact(update,x0=FloatSlider(description=r'\(x_0\)', min=0., max=1.,value=0.5),\
                phi0=FloatSlider(description=r'\(\varphi_0\)', min=-3., max=3.,value=0.),\
                omega0=FloatSlider(description=r'\(\omega_0\)', min=.5, max=3.,value=1.));
```

#### 11.1.4 Interpretation
<div class="alert alert-success">
$$x(t_1)=x(t_1+T_0)$$
</div>

  $\omega_0=\frac{2\pi}{T_0}$
  
  $\omega_0=2\pi\cdot f_0$

  $f_0 = \frac{1}{T_0}$

  $\omega_0$: angular frequency
  
  $f_0$: frequency

  $T_0$: period duration


<div class="alert alert-info">General differential equation of a free undamped oscillation
    
$$ \ddot{x}+\frac{k}{m}\,x = 0$$

can be written as

$$ \ddot{x}+\omega_0^2\,x = 0$$

with 
$$\omega_0^2=\frac{k}{m}$$.
</div>

Physical Pendulum:

$$\ddot{\varphi}+\frac{s\cdot m\cdot g}{J_A}\,\varphi = 0\quad\quad \omega_0^2=\frac{s\cdot m\cdot g}{J_A}$$


* A mathematical pendulum is a pendulum with a point mass attached to a massless string 

  <img src="figures/24.1_4.png" alt="24.1_4" style="width: 150px;"/>
 
  The moment of inertia
  
  $$J_A=m\cdot l^2$$ ⇨ $$\omega_0^2=\frac{s\cdot m\cdot g}{m \cdot l^2}$$ ⇨ $$\omega_0^2=\frac{g}{l}$$
  
<div class="alert alert-info">Reduced pendulum length: equivalent to the length $l$ in the equation of motion for the mathematical pendulum with the same oscillation period as the physical pendulum.
$$ l_{red}=\frac{J_A}{m\cdot s}$$</div>


#### 11.1.5 Energy balance for the spring oscillator

  \begin{eqnarray}
  E_{tot}&=&E_{kin}+E_{pot}\\
  &=& \underbrace{\frac{m}{2}\dot{x}^2}_{\text{kinetic and}} + \underbrace{\frac{k}{2}x^2}_{\text{potential energy}}
  \end{eqnarray}
  
  $$ \dot{x}=-x_0\,\omega_0\,\sin(\omega_0\,t)=-x_0\cdot\sqrt{\frac{k}{m}}\,\sin(\omega_0\,t)$$
  
  \begin{eqnarray}
  E_{tot}&=&\frac{k}{2}\,x_0^2\,\sin^2(\omega_0\,t)+\frac{k}{2}\,x_0^2\,\cos^2(\omega_0\,t)\\
  &=& \frac{k}{2}\,x_0^2\underbrace{\left[\sin^2(\omega_0\,t)+\cos^2(\omega_0\,t)\right]}_1
  \end{eqnarray}
  
  <div class="alert alert-info">The total energy is constant, thus it is not a function of time!
  $$E_{tot}=\frac{k}{2}\,x_0^2=\text{const.}$$
  
  </div>
