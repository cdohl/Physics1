## Derivation of apparent forces in rotating coordinate systems

<img src="figures/add1_KinematikKreisbewegung.png" width=150px>

The velocity vector for circular motion with angular velocity $\omega$ and rotational axis orientation $\frac{\vec{\omega}}{\omega}$ is given by

$$\vec{v}=\vec{\omega}\times\vec{r}\quad.$$

The rotation causes the vector $\vec{r}$ to change continuously. A co-rotating Cartesian coordinate system (drawn dashed here) with unit vectors ${\vec{e}_x}',{\vec{e}_y}',{\vec{e}_z}'$ or ${\vec{e}_i}'$ for short with $i=1,2,3$ also changes its direction. Its velocity $\frac{\mathrm{d}\vec{e}'_i}{\mathrm{d}t}=\dot{{\vec{e}_i}'}$ is therefore

$$\dot{{\vec{e}_i}'}=\vec{\omega}\times{\vec{e}_i}'\quad.\tag{1}$$

In general, a velocity of a vector in the inertial (non-rotating) system can thus $\vec{f}(t)=f_x\vec{e}_x+f_y\vec{e}_y+f_z\vec{e}_z$
 be written as:
 
$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}\vec{e}_x+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}\vec{e}_y+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}\vec{e}_z$$

However, if we describe the vector $f(t)$ with the rotating coordinate system, we must apply the chain rule, as the unit vectors also change over time.

$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}{\vec{e}_x}'+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}{\vec{e}_y}'+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}{\vec{e}_z}'+
f_x\frac{\mathrm{d}{\vec{e}_x}'}{\mathrm{d}t}+
f_y\frac{\mathrm{d}{\vec{e}_y}'}{\mathrm{d}t}+
f_z\frac{\mathrm{d}{\vec{e}_z}'}{\mathrm{d}t}$$

We can use the above equation (1) to transform the right-hand side:

$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} = 
\frac{\mathrm{d}\,f_x}{\mathrm{d}t}{\vec{e}_x}'+
\frac{\mathrm{d}\,f_y}{\mathrm{d}t}{\vec{e}_y}'+
\frac{\mathrm{d}\,f_z}{\mathrm{d}t}{\vec{e}_z}'+
\omega \times \left(\,
f_x\,{\vec{e}_x}'+ \,f_y\,{\vec{e}_y}'+ \,f_z\,{\vec{e}_z}'\right)$$

or this way
$$\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t} =
\left(\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t}\right)_r+\omega\times\vec{f}(t)$$
here $\left(\frac{\mathrm{d}\,\vec{f}}{\mathrm{d}t}\right)_r$ is the velocity measured in the rotating coordinate system.

$$\vec{v}_i=\left(\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}=\vec{v}_r+\omega\times\vec{r}$$

The acceleration is the second time derivative and can be formally expressed as

$$\vec{a}_i=\left(\frac{\mathrm{d}^2\vec{r}}{\mathrm{d}t^2}\right)_i=
\left(\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}\right)_i=
\left[\left(\frac{\mathrm{d}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}\right]
\left[\left(\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\right)_r+\omega\times\vec{r}\right]
$$

can be written. If we carry out the individual derivations (good exercise on the chain rule) you get:

$$\vec{a}_i=\left(\frac{\mathrm{d}^2\vec{r}}{\mathrm{d}t^2}\right)_r+2\vec{\omega}\times\vec{v}_r+\omega\times\left(\omega\times\vec{r}\right)+\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}\tag{2}$$

The first term on the right-hand side of Eq. (2) is the measured acceleration in the rotating coordinate system, i.e. $\vec{a}_r$. This allows us to compare this acceleration with the acceleration $\vec{a}_i$ in the inertial frame.

$$\vec{a}_r=\vec{a}_i-2\vec{\omega}\times\vec{v}_r-\omega\times\left(\omega\times\vec{r}\right)-\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}\tag{3}$$

This means that in addition to the acceleration $\vec{a}_i$, the following accelerations occur in the rotating coordinate system:

Centrifugal acceleration: $-\omega\times\left(\omega\times\vec{r}\right)$

Coriolis acceleration: $-2\vec{\omega}\times\vec{v}_r=2\vec{v}_r\times\vec{\omega}$ 

and the Euler acceleration: $-\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}$

These three additional accelerations correspond to the inertial forces


Centrifugal forece: $\vec{F}_{Zf}=-m\omega\times\left(\omega\times\vec{r}\right)$

Coriolis force: $\vec{F}_C=2\,m\,\vec{v}_r\times\vec{\omega}$ 

Euler force: $\vec{F}_{E}=-m\,\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}\times\vec{r}$
