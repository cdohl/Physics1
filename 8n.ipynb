## 8.1 Rotational dynamics
 
 Let us describe the rotatation of a disk with radius $R$ rotating around a center point C.
 
 The angle of of rotation is called $\Theta$, and the tangential velocity of the rim of the disk is $v=\omega R=\dot{\theta}R$.
 
<img src="figures/8n.1_1.png" width=240px>

 If we allow for an acclerated motion, i.e. the angular velocity is allowed to change, then we have a 
 $a_\mbox{tan}=\dot{\omega}R=\ddot{\Theta}R=\alpha R$.
 
 Here $\alpha$ is the angular acceleration.
 
 We can connect the linear and the circular motion by replacing the variables as shown in the table below.
 
| linear motion | circular motion |
| --- | ---|
| x |  $\Theta$ | 
| v |  $\omega$ | 
| a |  $\alpha$ | 
| m |  I | 


Thus the position $x$ in linear motion becomes the angle $\Theta$. We can rewrite the formulas for motion with constant acceleration first as linear motion:

$$x(t)=x_0+v_0 t + \frac{1}{2}a t^2$$

and for circular motion:

$$\Theta(t)=\Theta_0+\omega t + \frac{1}{2}\alpha t^2\quad .$$

And similiarily for the velocity in linear motion:

$$v(t)=v_0 + a t$$

compared to 

$$\omega(t)=\omega_0+\alpha t\quad .$$


## 8.2 Rotational energy and moment of inertia

Let us now describe the kinetic energy of a rotating disk. The disk rotates through a stationary axis C.


<img src="figures/8n.1_2.png" width=240px>


We start first with the kinetic energy of the mass element $m_i$ due to the rotation of the disk. This is straightforward:

$$K_i=\frac{1}{2}m_i v_i^2$$

Here the velocity $v_i$ is a function of the location on the disk. This can be simplified knowing that $v=\omega r_i$ to

$$K_i=\frac{1}{2}m_i \omega^2 r_i^2$$

The kinetic energy of the whole disk is then

$$K_\mbox{disk}=\frac{\omega^2}{2}\sum m_i r_i^2$$

The sum is named the moment of inertia of the rotating disk, i.e. 

$$I_C=\sum m_i r_i^2\quad .$$

Here $I_c$ stands for the moment of inertia of an object rotating through the axis C. This is important to remember, that the moment of inertia depends on the specific axis the object is rotating through.

We can  writt $I_C$ as an integral $I_C=\int_V r^2 dV$ and solved for. This is a mathematical excercise (which will do below) yet for now let us use the result for the disk that is

$$I_{C,\mbox{disk}}=\frac{1}{2}m R^2$$

## 8.3 Parallel axis theorem

<img src="figures/8n.1_5.png" width=250px>


The moment of inertia of an object rotating along axis $l$ going through the center of mass be $I_l$. If the object is rotated through an axis parallel to $l$ at a distance $d$, thus through $l'$ as shown in the figure above the moment of inertia can be obtained from the parallel axis theorem:

$$I_l'=I_l+m d^2\quad ,$$

where $m$ is the mass of the object.



## 8.4. Rotational and potential energy

Let us first have a look at the rolling motion

### Rolling motion
A rolling motion consists of the translation of the center of mass and a rotation around an axis through the center of gravity of the rolling body. In ideal rolling without sliding, both speeds are given by the relationship $v=\omega\,R$.


<img src="figures/8n.1_6.png" width=350px>


### Calculation of the moment of inertia for a cylinder (example)


A cylinder with constant density that rotates around the axis of symmetry $\vec{\omega}=\omega\,\vec{e}_z$ through the center $C: x=y=z=0$:

<img src="figures/8.3_2.png" width=200px>

$$I_{C,z}=\int r^2 \,\mathrm{d}m\quad .$$

At constant density, $\mathrm{d}m=\rho\,\mathrm{d}V$ and

$$I_{C,z}=\rho\int r^2 \,\mathrm{d}V\quad .$$

A suitable volume element must be used for the integration over $\mathrm{d}V$. Cylindrical coordinates are suitable here. These are easy to understand:

<img src="figures/8.3_3.png" width=250px>

The surface element $\mathrm{d}A$ of a cross-section of the cylinder is

$$\mathrm{d}A=\underbrace{r\,\mathrm{d}\varphi}_{\mathrm{arc length}}\,\mathrm{d}r$$

The volume then only takes into account the depth in the $z$-direction:

$$\mathrm{d}V=r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

$$I_{C,z}=\rho\int r^2 \,\mathrm{d}V\quad=
\rho\int_{-h/2}^{h/2}\int_0^R\int_0^{2\pi}r^2\,r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

Integration over $\mathrm{d}\varphi$ results in $2\pi$

$$I_{C,z}=2\pi\,\rho\int_{-h/2}^{h/2}\int_0^R r^3\mathrm{d}r\,\mathrm{d}z$$

Integration over $\mathrm{d}z$ results in $h$

$$I_{C,z}=2\pi\,\rho h \int_0^R r^3\mathrm{d}r$$

And finally via $r$

$$I_{C,z}=2\pi\,\rho h \left[\frac{r^4}{4}\right]_0^R=\underbrace{\rho \pi R^2 h}_{\rho\,V=m} \,\frac{1}{2}R^2=\frac{1}{2}m R^2$$



### Speed of a rolling cylinder
We can determine the speed of a cylinder rolling down an inclined plane using the law of conservation of energy. To do this, we assume that no frictional work is performed (the contact point between the cylinder and the plane is at rest).

<img src="figures/8.3_5.png" width=350px>


If the cylinder is at rest at the beginning, then 

$$E_{pot}=m g h$$

Then it speeds up. We have two kinetic energies due to the translation and the rotation:

$$E_{kin}=\frac{1}{2} m v^2+\frac{1}{2}I_C\,\omega^2$$

After the cylinder has rolled down the height $h$, we have converted the potential energy into kinetic energy, so the velocity is

$$v=\sqrt{\frac{2 g h}{1+\frac{I_C}{m R^2}}}$$

In the case of a cylinder with $I_C=\frac{1}{2}m R^2$

we obtain a velocity of 

$$v=\sqrt{\frac{4}{3}g\,h}<\sqrt{2 g h}$$

The velocity is less than the velocity of a mass point falling freely a height $h$, namely $v=\sqrt{2 g h}$. Why is this?



## 8.5 Angular momentum

<img src="figures/8.2_1.png" width=280px>

The angular momentum of a body with velocity $\vec{v}$ and mass $m$, and momentum $\vec{p}$ is given by

$$\vec{L}_A=\vec{r}\times(m\,\vec{v})=\vec{r}\times\vec{p}$$

Here $\vec{r}$ is the position vector to the reference point $A$. The angular momentum $\vec{L}$ is always dependent on the reference point, and therefore we write $\vec{L}_A$.

* Magnitude of $\vec{L}$

  $$\left|\vec{L}\right|=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin\left(\vec{r},\vec{p}\right)=\left|\vec{r}\right|\,\left|\vec{p}\right|\,\sin \gamma$$
  
  * $\left|\vec{L}\right|$ becomes maximum when $\vec{r}\perp\vec{v}$
  
  * $\left|\vec{L}\right|$ becomes minimum if $\gamma=0^\circ$ or $\gamma=90^\circ$, i.e. $\vec{r}\parallel\vec{v}$
  
  * Direction of $\vec{L}$ with the right-hand rule
  
  
* $\vec{L}_Z$ of a point mass on a circular path


  <img src="figures/8.2_2.png" width=240px>
  
  We select the center of the circle $Z$ as the reference point. Using the right-hand rule, you can see that the angular momentum is pointing towards you from the image plane. This means that the direction is identical to the direction of the axis of rotation $\vec{\omega}$. The magnitude is easy to determine as $\vec{r}_Z\perp\vec{v}$.

  \begin{eqnarray}
  \left|\vec{L}_Z\right|&=&m\,\left|\vec{r}_Z\times\vec{v}\right|\\
  &=&m\,r_Z\,v=m\,{r_Z}^2\,\omega
  \end{eqnarray}
  
  Since we know the direction of $\vec{L}_Z$ we can immediately use
  
  $$\vec{L}_Z=m\,{r_Z}^2\,\vec{\omega}$$

  $I_{A,\vec{\omega}}=m\,r^2$ is the moment of inertia with respect to an axis $\vec{\omega}$ through the point $A$, then the angular momentum is
  
  $$\vec{L}=I_{A,\vec{\omega}}\,\vec{\omega}$$

## 8.5 Basic law of rotation

* During translation, a force leads to a change in momentum

  $$ \vec{F}=\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}$$

* When does the angular momentum change?
  To do this, we determine the temporal change of $\vec{L}$:
  
  \begin{eqnarray}
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\frac{\mathrm{d}\,(\vec{r}\times\vec{p})}{\mathrm{d}t}\\
  &=&\underbrace{\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}\times\vec{p}}_{\vec{v}\times\vec{p}=0}+\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}\\
  \frac{\mathrm{d}\vec{L}}{\mathrm{d}t}&=&\vec{r}\times\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}=\vec{r}\times\vec{F}
  \end{eqnarray}

* The change in angular momentum over time is called torque $\vec{\tau}$

  $$\vec{\tau}=\vec{r}\times\vec{F}$$
  
  The angular momentum on a body is
  
  $$\vec{L}=I_A\,\vec{\omega}$$
  
  and thus is 
  
  $$\vec{\tau}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}=
  \frac{\mathrm{d}I_A}{\mathrm{d}t}\vec{\omega}+I_A\,\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$

  The angular acceleration is 
  
  $$\vec{\alpha}=\frac{\mathrm{d}\vec{\omega}}{\mathrm{d}t}$$
  
  and for the case of a constant moment of inertia we obtain the *basic law of rotation*
  
  $$\vec{\tau}=I_A\,\vec{\alpha}$$


## 8.6 Law of conservation of angular momentum

* closed physical system

  $$\vec{\tau}=\sum_{i=1}^N\vec{\tau}_i$$
  
  There is no external force acting, therefore $\vec{\tau}=0$ and since $\vec{\tau}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}$ is $\vec{L}=\mathrm{const}$.
  
* for several bodies in the system

  $$\vec{L}=\sum_{i=1}^N\vec{L}_i$$
  
  
* Example 1: $\vec{L}=0$

  for two bodies:
  
  $$\vec{L}=I_1\vec{\omega}_1+I_2\vec{\omega}_2=0$$
  
  $$\vec{\omega}_2=-\frac{I_1}{I_2} \vec{\omega}_1$$
  
  
* Example 2: $\vec{L}=\mathrm{const.}\ne0$

  Initial state
  $$\vec{L}= I_1\vec{\omega}_1$$
  
  Final state
  $$\vec{L}= I_2\vec{\omega}_2$$
  
  $$\vec{\omega_2}=\frac{I_1}{I_2}\vec{\omega}_1$$

## 8.7 Gyroscope: Free axes

* Previously: Rotation around fixed axes

* Now: Movement of rigid bodies without fixing the axis of rotation $\rightarrow$ Gyroscope.

We only consider rotations through the center of gravity. We call the symmetry axes of the object $\vec{a}$, $\vec{b}$ and $\vec{c}$.

In general, the gyroscope has a moment of inertia for each axis of symmetry.

The axis with the largest moment of inertia is $\vec{c}$. The one with the lowest $\vec{a}$: $I_a\le I_b\le I_c$

Symmetrical gyroscopes are gyroscopes in which at least 2 moments of inertia are equal.

<img src="figures/8.7_1n.png" width=500px>

* prolate gyroscope: $I_a<I_b=I_c$

* oblate gyroscope: $I_a=I_b<I_c$

The rotation parallel to $\vec{a}$ and $\vec{c}$ are stable.

## 8.8 Gyroscope: Precession

We consider a wheel that is mounted on an axis in the direction of the $x$-axis.

<img src="figures/8.8_1.png" width=200px>

The wheel is stationary and is now deflected by a torque $\vec{\tau}$ that acts for a short time $\Delta t$. We imagine the experiment in space without external forces. For example, the torque can be generated by a pair of forces that both act with the same force but in opposite directions on the wheel's center of gravity.

<img src="figures/8.8_2n.png" width=500px>

The wheel receives an angular momentum as

$$\vec{\tau}=\frac{\mathrm{d}\vec{L}}{\mathrm{d}t}\approx\frac{\Delta \vec{L}}{\Delta t}$$

and therefore rotates around the $z$-axis with angular momentum $L_z\vec{e}_z$.

#### With intrinsic angular momentum

Now the wheel is given an intrinsic angular momentum, i.e. an angular momentum around the center of gravity about the axis of symmetry. The wheel rotates around the axis at angular velocity $\omega$. Again, a torque acts on the wheel for a short time due to the force pair $\vec{F}$. Now the question arises, how does the wheel move after the time $\Delta t$?
The answer is not so simple. Because after the torque has acted, the angular momentum must be $\vec{L}=\mathrm{const.}$. This means that the wheel cannot simply rotate around the $z$-axis as above. Then the torque would continuously change its direction. The answer is described in the figure.

<img src="figures/8.8_3n.png" width=500px>

The torque $\vec{\tau}$ changes the angular momentum by the value $\vec{\Delta L}$, i.e. the axis of rotation is changed. After the time $\Delta t$, the new axis of rotation remains the same.


Let's illustrate this again using the second example. The wheel rotates with the same intrinsic angular momentum $\vec{L}$, but now a torque is briefly exerted in the direction of the negative $x$-axis, $\vec{\tau}$ as indicated in the figure below on the left.

<img src="figures/8.8_4n.png" width=500px>

The result is a change in the intrinsic angular momentum $\vec{L}$ in the direction of the torque, the wheel rotates briefly around the $z$-axis and adjusts the angular momentum axis to the new $\vec{L}$ direction.

Changing the direction of the angular momentum of a gyroscope is called *precession*. We can also quantitatively describe the change in the orientation of the gyroscope with a simple experiment. To do this, the spinning top is made to rotate and suspended from a rope so that it can rotate. The weight force $\vec{F_g}$ acts on the center of gravity of the gyroscope, generating a torque.

<img src="figures/8.8_5n.png" width=200px>

The torque is 

$$\vec{\tau}=\vec{PQ}\times \vec{F_g}$$

The torque $\vec{\tau}$ points into the image plane. The angle between $\vec{PQ}$ and $\vec{F_g}$ is $90^\circ$ and therefore the magnitude of

$$\left|\vec{\tau}\right|=r\,m\,g\quad .$$

The torque changes the angular momentum. The angular momentum follows the torque. The gyroscope rotates around the $z$-axis counterclockwise at a constant angular velocity of precession. 


The angular velocity of precession $\omega_{pr}$ is given here without derivation:

$$\omega_{pr}=\frac{\tau}{L_s}\quad,$$

where $L_s$ is the magnitude of the gyroscope's intrinsic angular momentum and $\tau$ is the magnitude of torque. This formula only applies if the intrinsic angular momentum is much greater than the angular momentum due to precession.

The torque is

$$\tau=r\,m\,g$$

and the angular momentum is

$$L_s=I_{\mathrm{wheel}}\,\omega\quad.$$

For a wheel in which the entire mass is distributed on a ring with radius $R$, $I_{\mathrm{wheel}}=m\,R^2$. $\omega$ is the angular velocity of the intrinsic angular momentum (the spin).

$$\omega_{pr}=\frac{r\,m\,g}{m\,R^2\,\omega}=\frac{r\,g}{R^2\,\omega}$$



## 8.9 Kepler's laws

### 8.9.1 Law of gravitation

<img src="figures/8.10_1.png" width=350px>

Masses attract each other. So far, we have used the gravitational force $\vec{F}_g=m\,\vec{g}$ and $\vec{g}$ the “constant” acceleration due to gravity. This is a simplification of the general law of gravity that describes the attraction of two masses $m_1$ and $m_2$:

$$\vec{F}_G=-\gamma\,\frac{m_1\,m_2}{r^2}\,\vec{e}_r\quad .\tag{8.1}$$

Here $r$ is the distance between the masses with distance vector $\vec{r}=r\,\vec{e}_r$ and $\gamma$ is the general gravitational constant:

$$\gamma=6.67408\pm0.00031\cdot 10^{-11}\mathrm{m}^3\mathrm{kg}^{-1}\mathrm{s}^{-2}\quad .$$

### 8.9.2 1st Kepler's law

The planets move on ellipses with the sun at one of their focal points.

```python
%matplotlib inline
%config InlineBackend.close_figures=False 

import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['font.size'] = 15

t = np.linspace(0., 2.*np.pi, 1000)
plt.figure(1, figsize=(8, 8));

def update_plot(a,b):
    e=np.sqrt(1-b**2/a**2)
    c=a*e
    ax=plt.gca()
    ax.clear()
    x=a*np.cos(t)
    y=b*np.sin(t)
    plt.plot(x,y)
    ## Foki
    plt.plot(c,0,'go',markersize=15,alpha=0.5)
    plt.plot(-c,0,'go',markersize=15,alpha=0.5)
    ax.arrow(0, 0, -c, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(-c, 0, c, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(-c/2,0.05,'f')
    
    ## semiaxis
    plt.plot([-a,a],[0,0],'k--',lw=1)
    plt.plot([0,0],[-b,b],'k--',lw=1)
    ax.arrow(0, 0, a, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(a, 0, -a, 0, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(0, 0, 0, b, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.arrow(0, b, 0, -b, lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(a/2,-0.1,'a')
    ax.text(-0.1,b/2,'b')
    
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
 
    plt.show();

w_a = widgets.FloatSlider(.8, min=.5, max =1., step=0.025, description=r'$a$:')
w_b = widgets.FloatSlider(.5, min=.2, max =0.5, step=0.025, description=r'$b$:')
widgets.interactive(update_plot, a=w_a, b=w_b)
```

### 8.9.3 Kepler's 2nd law

The radius vector (ray) from the sun to the planet sweeps over equal areas in equal times, see figure below.

<img src="figures/8.10_3.png" width=320px>

<img src="figures/8.10_4.png" width=270px>

$$\Delta A=\frac{1}{2}\,\left|\vec{r}\right|\,\left|\vec{v}\right|\Delta t\sin\beta$$
Now we let the interval $\Delta t\rightarrow 0$ go.

$$\frac{\mathrm{d}A}{\mathrm{d}t}=\frac{1}{2}\, \left|\vec{r}\right|\,\left|\vec{v}\right|\,\sin\beta$$

Insert the angular momentum:

$$\frac{\mathrm{d}\vec{A}}{\mathrm{d}t}=\frac{1}{2}\, \vec{r}\times\frac{1}{m}\vec{p}=\frac{1}{2 m} \vec{L}$$

Since the angular momentum is constant (there is no torque), the following applies

$$\frac{\mathrm{d}\vec{A}}{\mathrm{d}t}=\frac{1}{2 m} \vec{L}=\mathrm{const.}$$

### 8.9.4 3rd Kepler's law

The squares of the orbital periods of the planets behave like the third powers of their major semi-major axes.

$$\frac{T_1^2}{T_2^2}=\frac{a_1^3}{a_2^3}$$

or 

$$\frac{T_1^2}{a_1^3}=\frac{T_2^2}{a_2^3}=\dots=\mathrm{konst.}$$

We prove this for orbits around the sun with mass $m_S$. To do this, we consider two planets with mass $m_1, m_2$ and radii $r_1, r_2$ and orbital period $T_1, T_2$.

$$\vec{F}_1=-\gamma\frac{m_S\,m_1}{r_1^2}\vec{e}_r$$

On the circular path, the radial force is equal to the attractive force.

$$\vec{F}_{Rad,1}=-m_1\,\omega_1^2\,r_1\,\vec{e}_r$$


and thus

$$\gamma\frac{m_s}{r_1^2}=\omega_1^2\,r_1$$

Now we replace $\omega_1=2\pi/T_1$ and get

$$\frac{T_1^2}{r_1^3}=\underbrace{\quad 4\pi^2\frac{1}{\gamma \, m_S}\quad}_\mbox{Identical for all planets around the sun}$$


So it follows

$$\frac{T_1^2}{r_1^3}=\frac{T_2^2}{r_2^3}\quad .$$
