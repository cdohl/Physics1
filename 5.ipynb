# 5. Inertial Forces
## 5.1 Inertial system

The position of the mass point $m$ can be described in the coordinate system $\Sigma$ with the coordinates $(x,y,z)$ or in the coordinate system $\Sigma'$ with the coordinates $(x',y',z')$.

<img src="figures/4.3_5.png" width=250px>

Both coordinate systems move away from each other with the velocity of $\vec{v}_{rel}$.

<b>All non-accelerated coordinate systems are called inertial systems. Newton's laws apply here.</b> 

## 5.2 Linearly accelerated reference systems

i.e. $\sigma'$ is accelerated relative to $\sigma$ â‡’ $\vec{v}_{rel}\ne \mathrm{const.}$

<img src="figures/4.3_6.png" width=250px>

The acceleration is only along the $x$-direction.

$$x=x'+s\quad y=y'\quad z=z'$$

$$\ddot{x}=\ddot{x'}+\ddot{s}$$

$$\ddot{\vec{r}}=\ddot{\vec{r'}}+\ddot{\vec{s}}$$

$$\ddot{\vec{a}}=\ddot{\vec{a'}}+\ddot{\vec{s}}$$

$$m\ddot{\vec{a}}=m\ddot{\vec{a'}}+m\ddot{\vec{s}}$$

$$\vec{F}=\vec{F}'+m\,\ddot{\vec{s}}$$

$$\vec{F'}=\vec{F}\underbrace{-m\,\ddot{\vec{s}}}_{\vec{F}_T~\mathrm{inertial force}}$$

$$\vec{F}_T=-m \ddot{\vec{s}}$$

In the accelerated coordinate system, the inertial force acts in the opposite direction to the acceleration vector. This explains why we do not feel any force in free fall: the gravitational force and the inertial force cancel each other out exactly.

## 5.3 Rotating reference systems

### 5.3.1 Centrifugal force

<img src="figures/5.3_1.png" width=250px>

The centrifugal force represents the inertial force in a rotating reference system with $\vec{v}'=\mathrm{const.}$.

Then these forces act in the rotating reference system

$$\vec{F}_{r} + \vec{F}_{Zf} = 0$$

and thus

$$
\begin{align}
\vec{F}_{Zf}&=-\vec{F}_r\\
 &=-\left(-m\frac{v^2}{r}\vec{e}_r\right)\\
\end{align}
$$

Thus the centrifugal force is:

$$\vec{F}_{Zf}=m\,\frac{v^2}{r}\vec{e}_r$$

### 5.3.2 Coriolis force

If the relative motion in the rotating coordinate system is $\vec{v}\ne 0$, then an additional fiticious force acts, i.e. the Coriolis force.

$$\vec{F}_C=2\,m\left(\vec{v}'\times\vec{\omega}\right)$$

$\vec{\omega}$ is the vector of the angular velocity. The magnitude of $\vec{\omega}$ indicates the angular velocity and the direction indicates the orientation of the axis of rotation. This must be taken into account when determining the direction of the Coriolis force.

We first look at the Coriolis force on a disk. In the animation below, we are looking at a linear trajectory from an inertial system on a rotating disk. The lower view is from an observer who is on the disk, e.g. at the location marked with a red dot. There, the trajectory is curved and the Coriolis force acts for the observer on the object.

The Coriolis force also moves the pendulum bob of the *Focault pendulum*.

<img src="figures/5.3_4.png" width=220px>

From the drawing you can already see that the plane of the pendulum in the Northern Hemisphere rotates clockwise, the Coriolis force is maximum at the poles, and no Coriolis force acts on the pendulum at the equator.

It can be shown that the period for the rotation of Foucault's pendulum plane is 

$$ T=\frac{2\pi }{\Omega \,\sin(\varphi )}\quad ,$$

where $\Omega$ is the angular velocity of the earth and $\varphi$ is the latitude.

This means that in Germany, the plane rotates by approx. $11.5^\circ$ per hour.
